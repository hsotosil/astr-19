Import matplot lib and numpy.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Create an array x ranging from [0,1] in 100 steps and that defines a function that returns exp(x).

```python
x = np.arange(0, 1, 0.01)
y = 0.01 *x**2
```

Plot x vs. y and label the x-axis as “Time [milliseconds]” and the y-axis as “Awesomeness”.

```python
plt.plot(x,y)
plt.xlabel('Time [milliseconds]')
plt.ylabel('Awesomeness')

plt.savefig('cj2-7_exponential_plot.pdf',bbox_inches="tight",facecolor="white")
```
