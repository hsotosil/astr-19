Import matplotlib and numpy.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Using numpy.random.RandomState.uniform we can pull 1000 random numbers distrubted between the intergal [0,1]

```python
#Generate 1000 random numbers with a uniform distrubtion between [0,1].
n = np.random.uniform(0.0, 1.0, 1000)
np.all(n >= 0.0)
np.all(n < 1.0)
```

Histogram the random numbers into 100 bins.

```python
#Create the histogram bins
width = 0.5
histmin = np.floor(min(n))
histmax = np.ceil(max(n))+width

bins = np.arange(histmin, histmax, width)
plt.hist(n,bins=100, alpha=0.5, edgecolor="black")
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
plt.savefig("cj2-9_histogram_plot.pdf", bbox_inches="tight",facecolor="white")
```
